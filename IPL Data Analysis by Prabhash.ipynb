```python
#parsing the dates
matches['date'] =pd.to_datetime(matches.date)
matches['season'] = pd.DatetimeIndex(matches.date).year
matches['month'] = pd.DatetimeIndex(matches.date).month
matches['day'] = pd.DatetimeIndex(matches.date).day
matches['weekday'] =pd.DatetimeIndex(matches.date).weekday
```

```python
matches
```

```python
#since there are no missing values in venue we can use that to fill in the city
matches[matches['city'].isnull()]
```

```python
#judging by the venue all these matches have been played in Dubai
matches['city'].fillna('Dubai',inplace=True)
matches['city'].isnull().sum()
```

```python
matches[matches['player_of_match'].isnull()]
#there are null values for the winner, result and result_margin as well
#these rows seem to be invalid therefore we will remove them
```

```python
matches.drop([241,486,511,744], inplace=True)
matches.isnull().sum()
#now we only have missing values for result margin and method
```

```python
#dataframe containing null values in result_margin
matches[matches['result_margin'].isnull()]
```

```python
#since all the matches with null values in result_margin were tied, therefore the result_margin should be 0
matches['result_margin'].fillna(0, inplace=True)
```

```python
#I have observeed that the M Chinnaswamy Stadium has been mentioned in two different ways
#the observation made while plotting the pie chart
#also Rising Pune Supergiants had been mispelled in some case
matches['venue'].replace({'M.Chinnaswamy Stadium':'M Chinnaswamy Stadium'}, inplace=True)
matches.replace({'Rising Pune Supergiant':'Rising Pune Supergiants'}, inplace=True)
```

```python
#getting the unique values of the method column
matches['method'].unique()
```

```python
#there are lot of null values in the method column
#D/L which stands for Duckworth Lewis stern method which is used for rain affected matches only
#therefore we will fill 'regular' for all the other matches
matches['method'].fillna('regular', inplace=True)
```

```python
matches.isnull().sum()
#so all the missing values have been replaced
#also all the invalid rows have been removed
```

```python
#had osberved that RIsing Pune Supergiants had been mispelled at some places
deliveries.replace({'Rising Pune Supergiant':'Rising Pune Supergiants'}, inplace=True)
```

```python
#sum of missing values for each column
deliveries.isnull().sum()
```

```python
#number of wickets taken in all seasons
len(deliveries[deliveries.is_wicket==0])
```

```python
#merging the both datasets over id
full_record= matches.merge(deliveries, on="id")
full_record.head()
```

### Teams in team1 and team2

```python
matches_played_byteams=pd.concat([matches['team1'],matches['team2']],axis=1)
matches_played_byteams
```

### Matches played by each team

```python
teams=(matches_played_byteams['team1'].value_counts()+matches_played_byteams['team2'].value_counts()).reset_index()
teams.columns=['team_name','Matches_played']
teams
```

### Wins by each team

```python
wins=matches['winner'].value_counts().reset_index()
wins.columns=['team_name','wins']
wins
```

```python
player=teams.merge(wins,left_on='team_name',right_on='team_name',how='inner')
```

```python
player.columns=['team','matches_played','wins']
player
```

```python
player['%win']=(player['wins']/player['matches_played'])*100
player
```

```python
player.plot(kind='bar',x='team',figsize=(15,5))
plt.show()
```

### Runs pattern in different year

```python
match_years=full_record.groupby(['season'])['total_runs'].sum().reset_index()
match_years.set_index('season').plot()
plt.title('Total Runs Across the Seasons')
plt.show()
```

#### There was a decline in total runs from 2008 to 2009.But there after there was a  increase in runs in every season until 2013, but from next coming seasons there was a slump or irregularities in the total runs. But the number of matches are not equal in all seasons. We should check the average runs per match in each season

```python
match_played=matches.groupby(['season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_played
```

```python
match_years
```

```python
final=pd.concat([match_played,match_years.iloc[:,1]],axis=1)
final
```

### calculation of per match runs with a plot

```python
final['per_match_runs']=final['total_runs']/final['matches']
```

```python
final['per_match_runs'].plot()
```

### most lucky grounds for teams

```python
def lucky(matches,team_name):
    return matches[matches['winner']==team_name]['venue'].value_counts().nlargest(5)
```

```python
lucky(matches,'Mumbai Indians').plot(kind='bar')
```

### Which teams scored more than 200 runs

```python
high_scores=deliveries.groupby(['id', 'inning','batting_team','bowling_team'])['total_runs'].sum().reset_index() 
score_200=high_scores[high_scores['total_runs']>=200]
score_200.head(5)
```

### how many times a team scored 200+ runs

```python
topscorer_teams=score_200['batting_team'].value_counts()
topscorer_teams=pd.DataFrame(topscorer_teams)
topscorer_teams.style.background_gradient(cmap='PuBu')
```

### how many times a team conceded 200+ runs

```python
z=score_200['bowling_team'].value_counts()
z=pd.DataFrame(z)
z.style.background_gradient(cmap='PuBu')
```

### Most balls played by a batsman

```python
balls_played=deliveries.groupby(['batsman'])['ball'].count().reset_index()
balls_played.sort_values(by='ball',ascending=False).head(10).style.background_gradient(cmap='PuBu')
```

Virat kohli had played the most balls.
David Warner and Chris Gayle are the only two foreign players in this top 10 list.

### Top 10 run scorer of all time 

```python
runs=deliveries.groupby(['batsman'])['batsman_runs'].sum().reset_index()
runs.columns=['Batsman','runs']
y=runs.sort_values(by='runs',ascending=False).head(10).reset_index().drop('index',axis=1)
y.style.background_gradient(cmap='PuBu')
```

### Most number of 6's by any palyer

```python
six=deliveries[deliveries['batsman_runs']==6]['batsman'].value_counts().sort_values(ascending=False).head(10).reset_index()
six.plot(kind='bar',x='index',colormap='RdYlBu')
plt.show()
```

 ### Who has taken most wickets in all seasons and each individual season?
